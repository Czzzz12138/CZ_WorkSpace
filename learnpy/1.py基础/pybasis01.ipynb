## 1 注释与文档字符串
注释是用来解释代码的文本，Python中有两种类型的注释：单行注释和多行注释。
单行注释以#开头，后面跟随注释内容。例如：
```python
# 这是一个单行注释
print("Hello, World!")  # 这也是一个单行注释
```
文档字符串（docstring）是用来描述模块、函数、类或方法的字符串，通常放在定义的第一行。文档字符串使用三个双引号或三个单引号包围
也叫做多行注释，可以包含多行文本
多行注释可以使用三个单引号或三个双引号包围起来。例如：
```python
'''
这是一个多行注释
可以包含多行文本
'''

"""
这是另一个多行注释
也可以包含多行文本
"""
```
CTRL + / 可以注释或取消注释选中的行   
CTRL + Z 可以撤销上一步操作  
CTRL + F 可以查找  

## 2 print函数
print函数是Python中用来输出信息到控制台的函数。它可以接受多个参数，并将它们转换为字符串后输出。print函数的基本语法如下：
```python
print(*objects, sep=' ', end='\n', file=sys.stdout, flush=False)
```
- *objects：要输出的对象，可以是多个，用逗号分隔。
- sep：对象之间的分隔符，默认为空格。
- end：输出结束时的字符，默认为换行符。
- file：要写入的文件对象，默认为sys.stdout。
- flush：是否强制刷新输出缓冲区，默认为False。   
例如：
```python
print("Hello", "World", sep=", ", end="!\n")
```


````python
print("Hello", "World!",sep=", ",end="!!!\n")
````

## 3 变量
变量是用来存储数据的命名空间。   
在Python中，变量不需要声明类型，可以直接赋值。变量名必须以字母或下划线开头，后面可以跟字母、数字或下划线。   
### 变量的命名规则
- 变量名只能包含字母、数字和下划线，不能包含空格和特殊字符（如!、@、#等）；
- 变量名不能以数字开头；
- 变量名不能使用Python的保留字（如if、else、for等） ；   
     python的保留字可以通过以下代码查看：

````python
import keyword
print(keyword.kwlist)
````

- 变量名区分大小写，例如：myVar和myvar是两个不同的变量。
- 变量可以赋值为任何数据类型，例如：整数、浮点数、字符串、列表、字典等。
- 变量可以在程序的任何地方使用，但必须在使用之前赋值。


### 变量的命名规范
- 1.使用小写字母和下划线分隔单词（如 my_variable）
- 2.使用驼峰命名法（如 myVariable）大/小驼峰 
- 3.使用大写字母开头的类名（如 MyClass）
- 4.使用全大写字母的常量名（如 MY_CONSTANT）
- 5.变量名应该具有描述性，能够清楚地表达变量的含义

````python
name = "cz"
age = 18    
height = 1.70
print("Name:", name)
print("Age:", age)      
print("Height:", height)
````

### 数字类型
- 整数（int）：表示没有小数部分的数字，例如：1、-5、100等。
- 浮点数（float）：表示有小数部分的数字，例如：3.14、-0.001、2.0等。
- 复数（complex）：表示实数和虚数组成的数字，例如：1+2j、-3-4j等。
  
数字类型支持各种数学运算，例如：加法、减法、乘法、除法、取模等。可以使用内置的数学函数进行更复杂的计算，例如：abs()、round()、pow()等。   

````python
# 数字类型

# 整数类型  int
num1 = 10  
# 浮点数类型  float
num2 = 3.14
# 复数类型  complex
num3 = 1 + 2j
 
print("num1:", num1, "Type:", type(num1))
print("num2:", num2, "Type:", type(num2))
print("num1 + num2 =", num1 + num2, "Type:", type(num1 + num2))
````

### 字符串类型
字符串（str）是由一系列字符组成的文本数据类型。字符串可以使用单引号（'）或双引号（"）包围，例如：'Hello'、"World"等。   
字符串支持各种操作，例如：连接、切片、格式化等。可以使用内置的字符串方法进行更复杂的操作，例如：upper()、lower()、split()等。


````python
greeting = "Hello, World!"
message = '''
             Hello, cz!
             
             Welcome to Python programming.
          '''  # 三引号可以表示多行字符串
print("Greeting:", greeting)
print("Message:", message)
print("Greeting in uppercase:", greeting.upper())  # 转换为大写
print("Message in lowercase:", message.lower())  # 转换为小写
print("Greeting split:", greeting.split("l"))  # 分割字符串
````

### 布尔类型
布尔类型（bool）表示真（True）或假（False）的值。   
布尔值只有两个可能的值：True 和 False (注意首字母大写)。   
布尔值可以当作数字使用，其中 True 等价于 1，False 等价于 0。       
布尔值可以通过比较运算符（如 ==、!=、>、< 等）来生成。   
布尔值也可以通过逻辑运算符（如 and、or、not）来组合。    
布尔类型常用于条件判断和逻辑运算，例如：if语句、while循环等。可以使用内置的布尔函数进行更复杂的操作，例如：bool()、not、and、or等。

````python
print("True as int:", int(True))   
print("False as int:", int(False)) 
print("True + 1:", True + 1)       #这里运算时True被当作1处理
print("False + 1:", False + 1)     
````

### 5 类型转换
Python提供了内置的类型转换函数，可以将一种数据类型转换为另一种数据
- int() 函数可以将字符串或浮点数转换为整数
- float() 函数可以将整数或字符串转换为浮点数
- str() 函数可以将整数或浮点数转换为字符串  


````python
num_str = "123"  # 字符串类型
num_int = int(num_str)  # 转换为整数
num_float = float(num_str)  # 转换为浮点数
tuple(num_str)  # 转换为元组（每个字符为一个元素）
list(num_str)  # 转换为列表（每个字符为一个元素）  list(可迭代对象)
chr(num_int)  # 转换为字符（ASCII码对应的字符）
print("String to int:", num_int)  # 输出: String to int: 123
print("String to float:", num_float)  # 输出: String to float: 123.0
print("String to tuple:", tuple(num_str))  # 输出: String to tuple: ('1', '2', '3')
print("String to list:", list(num_str))  # 输出: String to list: ['1', '2', '3']
print("String to char:", chr(123))  # 输出: String to char: {对应的字符}
````

### 6 占位符与格式化输出

占位符是一种特殊的符号，用于在字符串中表示变量的位置   
Python 中常用的占位符有 %s、%d、%f 等  
- %s 用于表示字符串  
- %d 用于表示整数  
- %f 用于表示浮点数  
- %% 用于表示百分号  

占位符可以通过 % 运算符与变量结合使用  

````python
print("Name: %s" % name)  # 输出: Name: cz

print("Age: %d" % age)    # 输出: Age: 18
#%4d 表示整数至少占4个字符宽度，不足的部分用空格补齐，右对齐
print("Age: %4d" % age)   # 输出: Age:   18 (前面有两个空格)
#%-4d 表示整数至少占4个字符宽度，不足的部分用空格补齐，左对齐
print("Age: %-4d" % age)  # 输出: Age: 18   (后面有两个空格)
#04d 表示整数至少占4个字符宽度，不足的部分用0补齐，右对齐
print("Age: %04d" % age)  # 输出: Age: 0018 (前面有两个0)

print("Height: %f" % height)  # 输出: Height: 1.700000 默认6位小数
# 可以指定浮点数的精度
print("Height: %.2f" % height)  # 输出: Height: 1.70(保留两位小数 五舍六入)
# 也可以同时使用多个占位符
print("Name: %s, Age: %d, Height: %.2f" % (name, age, height))
# 输出: Name: cz, Age: 18, Height: 1.70
````
